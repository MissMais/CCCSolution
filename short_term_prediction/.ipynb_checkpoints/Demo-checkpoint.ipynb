```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
from math import sqrt
```

```python
ticker = 'TATAMOTORS.NS'
start_date = '2018-01-01'
end_date = '2023-10-31'
```

```python
df = yf.download(ticker, start_date, end_date).reset_index()
df
```

```python
close = df['Close']
close
```

```python
train_data = close.iloc[:int(len(close)*0.8)]
test_data = close.iloc[int(len(close)*0.8):]
train_data.shape, test_data.shape
```

```python
test_data.iloc[20]
```

```python
def data_preparation(data):
    X = []
    y = []
    for i in range(20, len(data)):
        X.append(data.iloc[i-20:i])
        y.append(data.iloc[i])
    return np.array(X), np.array(y)
```

```python
X_train, y_train = data_preparation(train_data)
X_test, y_test = data_preparation(test_data)
```

```python
X_train, y_train
```

```python
X_train.shape, y_train.shape
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, batch_size=200, epochs=500, verbose=1)
```

```python
pred_train = model.predict(X_train)
```

```python
pred_train
```

```python
mae = mean_absolute_error(y_train, pred_train)
print('mean absolute error-', mae)
mse = mean_squared_error(y_train, pred_train)
print('mean squared error-', mse)
```

```python
plt.plot(pred_train, color='red')
plt.plot(y_train, color='blue')
```

```python
pred_test = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, pred_test)
print('mean absolute error-', mae)
mse = mean_squared_error(y_test, pred_test)
print('mean squared error-', mse)
```

```python
plt.plot(pred_test, color='green')
plt.plot(y_test, color='orange')
```
